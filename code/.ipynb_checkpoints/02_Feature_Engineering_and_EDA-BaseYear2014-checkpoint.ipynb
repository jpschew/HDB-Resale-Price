## Add the nearest amenities to each of the HDB data
This section adds all the nearest amenities to the respective HDB data as well as computing the distance to the respective amenities. Longitude and latitude data were used to compute the distance between them using the Haversine formula.

```python
import datetime
def get_nearest_amenity_dist(df, amenity_df, amenity):
#     amenity_list = []
#     dist_list = []
    a = datetime.datetime.now()
    if f'nearest_dist_to_{amenity}' not in df:
        df[f'nearest_dist_to_{amenity}'] = ''
    if f'nearest_{amenity}' not in df:
        df[f'nearest_{amenity}'] = ''
    for index, row in df.iterrows():
        amenity_list = []
        dist_list = []
#     for hdb_lon, hdb_lat in zip(df['longitude']. df['latitude']):
        lat1 = row['latitude']
        lon1 = row['longitude']
        for amenity_lon, amenity_lat, amenity_name in \
                                    zip(amenity_df[f'{amenity}_long'], amenity_df[f'{amenity}_lat'], amenity_df[amenity]):
            lat2 = amenity_lat
            lon2 = amenity_lon
            dist = distance(lat1, lat2, lon1, lon2)
#             print(dist)
#             print(amenity_name)
            amenity_list.append(amenity_name)
#             print(amenity_list)
            dist_list.append(dist)
#             print(dist_list)
#             break
        
        
        amenity_dist_dict = dict(zip(amenity_list, dist_list))
#         print(amenity_dist_dict)
#         break
#         print(amenity_list, dist_list)
#         print(amenity_dist_dict.values())
#         break
        
#         # get minimum dist
        min_value = min(amenity_dist_dict.values())
#         print(min_value)
#         break
        df[f'nearest_dist_to_{amenity}'][index] = min_value
        # get amenity name
        for key, value in amenity_dist_dict.items():
            if value == min_value:
                df[f'nearest_{amenity}'][index] = key
                
        if (index+1) % 1000 == 0:
            b = datetime.datetime.now()
            print(f'{(index+1)} shortest distance to {amenity} have been added for  {b-a} seconds.')
    c = datetime.datetime.now()
    print(f'Total time taken to add {len(df)} data is {c-a} seconds.')
#         print(df)
#         break
#         df[f'nearest_{amenity}'][row] = {key for key, value in amenity_dist_dict.items() if value == min_value}
```

```python
# function to calculate Distance Between Two Points on Earth
# Distance, d = radius * arccos[(sin(lat1) * sin(lat2)) + cos(lat1) * cos(lat2) * cos(long2 – long1)]
from math import radians, cos, sin, asin, sqrt, acos
def distance(lat1, lat2, lon1, lon2):
     
    # The math module contains a function named
    # radians which converts from degrees to radians.
    lon1 = radians(lon1)
    lon2 = radians(lon2)
    lat1 = radians(lat1)
    lat2 = radians(lat2)
      
    # Haversine formula
    a = sin(lat1) * sin(lat2) + cos(lat1) * cos(lat2) * cos(lon2 - lon1)
 
    c = acos(a)
    
    # Radius of earth in kilometers. Use 3963 for miles or 6378.8 for km
    r = 6378.8
      
    # calculate the result
    return(c * r)
```

```python
data_df['latitude'] = data_df['latitude'].astype(float)
data_df['longitude'] = data_df['longitude'].astype(float)
mrt_df['mrt_lat'] = mrt_df['mrt_lat'].astype(float)
mrt_df['mrt_long'] = mrt_df['mrt_long'].astype(float)
pri_sch_df['pri_sch_lat'] = pri_sch_df['pri_sch_lat'].astype(float)
pri_sch_df['pri_sch_long'] = pri_sch_df['pri_sch_long'].astype(float)
sec_sch_df['sec_sch_lat'] = sec_sch_df['sec_sch_lat'].astype(float)
sec_sch_df['sec_sch_long'] = sec_sch_df['sec_sch_long'].astype(float)
```

## Feature Engineering
Feature engineerging was done in this section to add in the age of the flat as well as converting the 'storey_range' feature to the average storey of the flat. In this section, inflation rate was computed using the median resale price and wasalso  based on year 2014 so that all the resale price were converted to the resale price in 2014 for modeling purpose in the next part of the notebook. The base year was chosen to be 2014 as in the next part of the notebook, we will need to predict teh resale price of a HDB flat given a number of characteristics/features.

```python
def feature_eng(df):
    df['age'] = df['year'] - df['lease_commence_date']
#     df['psqm'] = round(df['resale_price']/df['floor_area_sqm'], 2)
    df['avg_floor'] = df['storey_range'].map({'01 TO 03': 2, '01 TO 05': 3, '04 TO 06': 5, '06 TO 10': 8, \
                                              '07 TO 09': 8, '10 TO 12': 11, '11 TO 15': 13, '13 TO 15': 14, \
                                              '16 TO 18': 17, '16 TO 20': 18, '19 TO 21': 20, '21 TO 25': 23, \
                                              '22 TO 24': 23, '25 TO 27': 26, '26 TO 30': 28, '28 TO 30': 29, \
                                              '31 TO 33': 32, '31 TO 35': 33, '34 TO 36': 35, '36 TO 40': 38, \
                                              '37 TO 39': 38, '40 TO 42': 41, '43 TO 45': 44, '46 TO 48': 47, \
                                              '49 TO 51': 50})
```

```python
def get_median_resale_price(df, year_list):
    median_resale = {}
    for year in year_list:   
        median_resale[year] = df.loc[df['year'] == year, 'resale_price'].median()
    return median_resale
```

```python
year_list = list(range(2012, 2023))
# year_list
```

```python
median_resale_price = get_median_resale_price(data_df, year_list)
median_resale_price
```

```python
def cal_inflated_rate(price_dict, base_year):
    inflated_rate = {}
    for year, price in price_dict.items():
        inflated_rate[year] = price_dict[base_year] / price
    return inflated_rate
```

```python
inflated_rate = cal_inflated_rate(median_resale_price, 2014)
inflated_rate
```

```python
def add_inflated_resale(df, inflated_rate_dict):
    if 'inflated_rate' not in df:
        df['inflated_rate'] = ''
    if 'inflated_resale_price' not in df:
        df['inflated_resale_price'] = ''
    for year, inflated_rate in inflated_rate_dict.items():
#         df['inflated_resale_price'] = df.loc[df['year'] == year, 'resale_price']*inflated_rate
        df.loc[df['year'] == year, 'inflated_resale_price'] = round(df.loc[df['year'] == year, \
                                                                           'resale_price']*inflated_rate, 2)
        df.loc[df['year'] == year, 'inflated_rate'] = inflated_rate
    return df
```

```python
data_df = add_inflated_resale(data_df, inflated_rate)
```

```python
def create_pairplot(df, x_cols, yvalue):
    sns.pairplot(
        data=df,
        x_vars=x_cols,
        y_vars=yvalue
    )
```

```python
def create_heatmap(df, cm=True):
    
    if cm:
        plt.figure(figsize=(12, 12))
        heat_map = sns.heatmap(
            df.corr()[['inflated_resale_price']].sort_values('inflated_resale_price', ascending=False), 
            annot=True, 
            fmt='.2g', 
            vmin=-1, vmax=1, center=0, 
            cmap='coolwarm',
            linecolor='black', linewidth=2,
            #square=True, mask=np.triu(df.corr())
        )
        heat_map.set_title(
                    'Correlation Matrix \nContinuous Variables \nVs \nInflated Resale Price', 
                    fontdict= { 'fontsize': 20, 'fontweight':'bold'}
        );
    else:
        plt.figure(figsize=(14, 14))
        heat_map = sns.heatmap(
            df.corr(), 
            annot=True, 
            fmt='.2g', 
            vmin=-1, vmax=1, center=0, 
            cmap='coolwarm',
            linecolor='black', linewidth=2,
            square=True, mask=np.triu(df.corr())
        )
        heat_map.set_title(
                    'Correlation Matrix \nbetween \nContinuous Variables ', 
                    fontdict= { 'fontsize': 20, 'fontweight':'bold'}
        );
```

```python
def create_boxplot(df, list_columns, target):
    rows = round(len(list_columns)/1)
    fig, ax = plt.subplots(rows, 1, figsize=(30, 75))
    ax = ax.ravel()
    for idx, feature in enumerate(list_columns):
        # get the index(key) to access the median; sorted by median score in desc order
        my_order = df.groupby(feature)[target].median().sort_values(ascending=False).index
    #             print(feature)
        
        bp = sns.boxplot(
            data=df,
            x=feature,#target,
            y=target,#feature,
#             hue=feature,
            ax=ax[idx],
            order=my_order
        )
        bp.set_xlabel(feature, fontdict={'fontsize': 16, 'fontweight': 'bold'})
        bp.set_ylabel('Mediam Resale Price', fontdict={'fontsize': 16, 'fontweight': 'bold'})
        bp.set_title(f'Median Resale Price vs {feature}', fontdict={'fontsize': 16, 'fontweight': 'bold'})
        bp.set_xticklabels(bp.get_xticklabels(), rotation=90)
#         plt.xticks(rotation=90)
    if len(list_columns) % 3 != 0:
        ax.flat[-1].set_visible(False)
```
