## Web-scraping of Location data using OneMap API

This section scraps in the location data such as the longitude and latitude for the HDB flat using OneMap API. These data were added to the HDB dataset which will be needed to be used later to compute the distance to the nearest amenities like mrt stations, primary and secondary schools in the next part of the notebook.

```python
import requests
```

```python
test_data = merge_data_drop.loc[:,:]
```

```python
test_data.head(10)
```

```python
test_data.tail(10)
```

```python
test_data.info()
```

```python
import datetime
def scrap_hdb_data(df):
    cont_scrap = True
    a = datetime.datetime.now()
    
    if 'longitude' not in df:
        df['longitude'] = ''
    if 'latitude' not in df:
        df['latitude'] = ''

#     while cont_scrap:
    for index, row in df.iterrows():
        query = row['street_name'] + ' ' + row['block']
    #     print(query)
        url = f'https://developers.onemap.sg/commonapi/search?searchVal={query}&returnGeom=Y&getAddrDetails=Y'
    #     print(url)
        res = requests.get(url)
        if res.status_code != 200:
            print('error code')
            break
    #     print(res)
        data = res.json()
#         print(data['results'][0])
    #     break
        try:
            df['longitude'][index] = data['results'][0]['LONGITUDE']
            df['latitude'][index] = data['results'][0]['LATITUDE']
        except:
            pass
        
        if (index+1) % 1000 == 0:
            b = datetime.datetime.now()
            print(f'{(index+1)} hdb data have been added longitude and latitude values for {b-a} seconds.')
    c = datetime.datetime.now()
    print(f'Total time taken to scrap {len(df)} hdb data is {c-a} seconds.')
    return df
```

## Web-scraping of amenities like MRT stations, Primary and Secondary Schools
This section scraps in the location data of all the MRT stations, primary and secondary schools in Singapore and saved to their respective dataframes. These data were needed to add the respective amenities that were nearest to all the HDB flats in the dataset. Some data cleaning were done in this section to remove duplicate mrt stations and schools.

```python
# var_list in the sequnce of names, long and lat
# amenity is the string to be queried
def scrap_amenity_data(amenity, var_list):
    # get mrt
    to_scrap = True
    page_num = 1
    var_list[0] = []
    var_list[1] = []
    var_list[2] = []
    while to_scrap:
        query = amenity
    #     print(query)
        url = f'https://developers.onemap.sg/commonapi/search?searchVal={query}&returnGeom=Y&getAddrDetails=Y&pageNum={page_num}'
    #     print(url)
        res = requests.get(url)
        if res.status_code != 200:
            break
    #     print(res)
        data = res.json()
        print(f'{page_num} page(s) retrieved')
    #     print(data['results'])
    #     for page in range(page_num, data['totalNumPages']+1):
    #     break
        for info in data['results']:
            var_list[0].append(info['BUILDING'])
            var_list[1].append(info['LONGITUDE'])
            var_list[2].append(info['LATITUDE'])
    #             print(mrt_names, mrt_long, mrt_lat)
    #             break
        page_num += 1
        if page_num > data['totalNumPages']:
            to_scrap = False
    # each page 10 mrt, total 91 pages, <=910 mrt
    print(f'The total number of {amenity} retrieved is {len(var_list[0])}') 
    return var_list
```

```python
def list_to_rm(df, var_list, amenity):
    for string in var_list:
        df[amenity] = df[amenity].str.replace(string, '')
```

```python
def check_amenity(df, amenity):
    amenity_dict = {'mrt': 'mrt stations', 'pri_sch': 'primary schools', 'sec_sch': 'secondary schools'}
    amenity_sorted = df[amenity].tolist()
    amenity_sorted.sort()
    print(amenity_sorted)
    print(f'The number of {amenity_dict[amenity]} is {len(df[amenity])}.')
```

```python
%time
# # series.str.replace() uses regular expression
# # . means any char except new line
# # means 0 or more instances of pattern from the left
# string_to_rm = ['NTUC', 'OCBC', 'UOB', 'DBS', 'MRT CIRCLE LINE STATION', 'NEL', 'MRT STATION.*']
# for string in string_to_rm:
#     mrt_df['mrt'] = mrt_df['mrt'].str.replace(string, '')

mrt = scrap_amenity_data('mrt station', ['mrt', 'mrt_long', 'mrt_lat'])
```
